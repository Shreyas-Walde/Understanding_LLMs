```python
input_tokens = tokenizer("Write a code for adding two numbers", return_tensors="pt") #convert user's input into token
input_tokens
```

```python
import torch
from transformers import AutoModelForCausalLM          # loading model

model = AutoModelForCausalLM.from_pretrained("google/gemma-3-1b-it", torch_dtype=torch.bfloat16)
```

```python
out = model(input_ids=input_tokens["input_ids"])
out
```

```python
gen_out = model.generate(input_ids=input_tokens["input_ids"], max_new_tokens=100)  # where the probability is high!
gen_out                                                        # it's not readable since it's tokens
                                                               # Model gave us response to user's input in token
```

```python
# detokenizing tokens
# detokenizing = making it readable to humans
tokenizer.batch_decode(gen_out)  #depends on token size
# for increasing token size we can use
# ,max_new_tokens=100
# Below is (vector embedding) in text form.
```
